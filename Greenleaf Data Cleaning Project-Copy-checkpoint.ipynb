# FINDING & HANDLING OUTLIERS WITH INTERQUARTILE RANGE (IQR)

def find_outliers_iqr(df, column):
    # Calculate Q1 and Q3
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    
    # Compute IQR
    IQR = Q3 - Q1
    
    # Define bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identify outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    
    return outliers, lower_bound, upper_bound



```python
# FINDING & HANDLING OUTLIERS WITH INTERQUARTILE RANGES:

def find_outliers_iqr(df, column):
    # Calculate Q1 and Q3
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    
    # Compute IQR
    IQR = Q3 - Q1
    
    # Define bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identify outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    
    return outliers, Q1, Q3, IQR, lower_bound, upper_bound
```

```python
outliers, Q1, Q3, IQR, lb, ub = find_outliers_iqr(
    greenleaf_sales,
    "Unit Price"
)
```

### COMPUTE CALCULATED COLUMNS USING Quantity Sold, Unit Price, Unit Cost

##### NEW COLUMNS: TOTAL REVENUE, TOTAL COST, TOTAL PROFIT

**1. Total Revenue**:

CALCULATED AS: Quantity Sold * Unit Price

**2. Total Cost**:

CALCULATED AS: Quantity Sold * Unit Cost

**3. Total Profit**:

CALCULATED AS: Total Revenue - Total Profit


```python
# COMPUTE CALCULATED COLUMNS USING Quantity Sold, Unit Price, Unit Cost

greenleaf_sales['Total Revenue'] = greenleaf_sales['Quantity Sold'] * greenleaf_sales['Unit Price']
```

```python
# COMPUTE CALCULATED COLUMNS USING Quantity Sold, Unit Price, Unit Cost

greenleaf_sales['Total Cost'] = greenleaf_sales['Quantity Sold'] * greenleaf_sales['Unit Cost']
```

```python
# COMPUTE CALCULATED COLUMNS USING Quantity Sold, Unit Price, Unit Cost

greenleaf_sales['Total Profit'] = greenleaf_sales['Total Revenue'] - greenleaf_sales['Total Cost']
```
